```python
c = wmi.WMI("localhost")
```

```python
# CPU
for cpu in c.Win32_Processor():
    print(cpu.LoadPercentage)

# RAM
os = c.Win32_OperatingSystem()[0]
print(int(os.FreePhysicalMemory) / int(os.TotalVisibleMemorySize) * 100)

# Dysk
for disk in c.Win32_LogicalDisk(DriveType=3):
    print(disk.DeviceID, disk.FreeSpace)

# Procesy
for p in c.Win32_Process():
    print(p.Name)
```

```python
import wmi
from datetime import datetime, timedelta

c = wmi.WMI("localhost")

# ostatnia godzina
dt = (datetime.now() - timedelta(hours=12)).strftime("%Y%m%d%H%M%S.000000-000")

query = f"""
SELECT TimeGenerated, SourceName, EventCode, Message
FROM Win32_NTLogEvent
WHERE Logfile = 'System'
AND Type = 'Error'
AND TimeGenerated > '{dt}'
"""

logs = list(c.query(query))
logs.sort(key=lambda x: x.TimeGenerated)

for log in logs:
    print(
        log.TimeGenerated,
        log.SourceName,
        log.EventCode,
        (log.Message or "")[:200]  # limit długości
    )
```

```python
import wmi

c = wmi.WMI("localhost")

query = """
SELECT TimeGenerated, SourceName, EventCode, Message
FROM Win32_NTLogEvent
WHERE Logfile = 'System'
AND SourceName = 'Service Control Manager'
AND (EventCode = '7000' OR EventCode = '7031' OR EventCode = '7034')
"""

for log in c.query(query):
    print(log.TimeGenerated, log.EventCode)
    print(log.Message)
    print("------")
```

```python
import wmi

w = wmi.WMI(namespace="root\\wmi")

# aktualna maksymalna pojemność
full = w.BatteryFullChargedCapacity()[0].FullChargedCapacity

# fabryczna pojemność
design = w.BatteryStaticData()[0].DesignedCapacity

health = (full / design) * 100

print(f"Full: {full}")
print(f"Design: {design}")
print(f"Health: {health:.2f}%")
```
