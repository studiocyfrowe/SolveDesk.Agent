```python
import pandas as pd
import wmi
```

```python
# Load Data
HOST_NAME = 'localhost'
data_source = wmi.WMI(f'{HOST_NAME}')
```

```python
import time
import pandas as pd

processor_hst = []

i = 0

while i <= 20:
    for cpu in data_source.Win32_Processor():
        new_data = {
            "Caption": cpu.Caption,
            "Description": cpu.Description,
            "LoadPercentage": cpu.LoadPercentage,
            "MaxClockSpeed": cpu.MaxClockSpeed
        }

        processor_hst.append(new_data)

    time.sleep(1)
    i += 1

processor_df = pd.DataFrame(processor_hst)
```
